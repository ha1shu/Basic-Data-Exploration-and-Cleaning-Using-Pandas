## Step 3 — Handle Missing Values




```python
missing       = df_raw.isnull().sum()
missing_pct   = (missing / len(df_raw) * 100).round(2)

missing_report = pd.DataFrame({
    'Missing Count' : missing,
    'Missing %'     : missing_pct,
    'Status'        : missing.apply(lambda x: ' Complete' if x == 0 else f' {x} missing')
}).sort_values('Missing Count', ascending=False)

print("── Missing Value Report ──────────────────────────────────────────────")
display(missing_report)
print(f"\nTotal missing cells : {missing.sum():,}  out of  {df_raw.size:,}  ({missing.sum()/df_raw.size*100:.2f}%)")
```

```python
df = df_raw.copy()   

# ── Fill numeric ─────────────
df['discount'] = df['discount'].fillna(0.0)

# ── Fill categorical / text ──────────────
df['seller_name']        = df['seller_name'].fillna('Unknown Seller')
df['seller_information'] = df['seller_information'].fillna('Not Available')
df['what_customers_said']= df['what_customers_said'].fillna('No Reviews')
df['variations']         = df['variations'].fillna('[]')

# ── Drop column with >78% missing ─────────
df.drop(columns=['videos'], inplace=True)

# ── Verify ──────────
remaining = df.isnull().sum().sum()
print(f"   Missing values remaining : {remaining}")
print(f"   Columns after drop        : {df.shape[1]}  (was {df_raw.shape[1]})")
```

## Step 5 — Remove Duplicates

Duplicate rows skew counts, totals, and machine-learning models.  
`drop_duplicates()` keeps the first occurrence of each repeated row and discards the rest.


```python

n_before = len(df)
n_dupes  = df.duplicated().sum()

print(f"Rows before deduplication : {n_before:,}")
print(f"Duplicate rows found      : {n_dupes:,}")

df = df.drop_duplicates().reset_index(drop=True)

n_after = len(df)
print(f"\nRows after deduplication : {n_after:,}  (removed: {n_before - n_after:,})")
```

##  Step 6 — Create Derived Column: `total_amount`

### Formula
$$\text{total\_amount} = \text{initial\_price} \times \text{quantity}$$

> **Note on `quantity`:** The Myntra dataset records product listings, not individual transactions — so there is no `quantity` column in the raw data.  
> For this assignment we derive a **realistic `quantity` estimate** by bucketing `ratings_count` (number of people who rated the product is a strong proxy for units sold):
>
> | ratings_count | estimated quantity |
> |---|---|
> | < 10 | 1 |
> | 10 – 49 | 2 |
> | 50 – 199 | 3 |
> | 200 – 499 | 5 |
> | ≥ 500 | 10 |
>
> This is a transparent, assignment-appropriate approach when quantity is not directly recorded.


```python
# ── Step 6a: Create 'quantity' from ratings_count bucket ─────────────────────
def estimate_quantity(rc):
    if rc < 10:   return 1
    if rc < 50:   return 2
    if rc < 200:  return 3
    if rc < 500:  return 5
    return 10

df['quantity'] = df['ratings_count'].apply(estimate_quantity)

print("Quantity distribution:")
print(df['quantity'].value_counts().sort_index().to_string())
```

```python
# ── Step 6b: Create total_amount = initial_price × quantity ──────────────────
df['total_amount'] = df['initial_price'] * df['quantity']

print("'total_amount' column created  (initial_price × quantity)")
print()
print(f" Min    : ₹{df['total_amount'].min():>10,.2f}")
print(f" Mean   : ₹{df['total_amount'].mean():>10,.2f}")
print(f" Median : ₹{df['total_amount'].median():>10,.2f}")
print(f" Max    : ₹{df['total_amount'].max():>10,.2f}")
print()
df[['title', 'category', 'initial_price', 'quantity', 'total_amount']].head(10)
```

```python
# ── Step 6c: Also derive discounted_amount ───────────────────────────────────
df['discounted_price']  = (df['initial_price'] * (1 - df['discount'] / 100)).round(2)
df['savings_per_unit']  = (df['initial_price'] - df['discounted_price']).round(2)

print(" Bonus derived columns: 'discounted_price' and 'savings_per_unit'")
df[['title', 'initial_price', 'discount', 'discounted_price', 'savings_per_unit']].head(6)
```
